#### Criando a dimensão calendário

```python

def calendario(ano, mes):
    # Obter o primeiro dia do mês
    primeiro_dia = datetime(ano, mes, 1)
    
    # Obter o último dia do mês
    ultimo_dia = primeiro_dia.replace(
        day=pd.Timestamp(ano, mes, 1).days_in_month
    )
    
    # Gerar as datas do mês
    datas_mes = pd.date_range(primeiro_dia, '2022-12-31', freq='D')
    
    # Criar DataFrame com as datas do mês
    calendario_df = pd.DataFrame({'Data': datas_mes})
    
    # Adicionar coluna com o dia da semana (0 = segunda-feira, 1 = terça-feira, ..., 6 = domingo)
    calendario_df['Dia_Semana'] = calendario_df['Data'].dt.dayofweek
    
    # Adicionar coluna com o número da semana no ano
    calendario_df['Semana_Ano'] = calendario_df['Data'].dt.isocalendar().week
    
    # Adicionar coluna com o dia do ano
    calendario_df['Dia_Ano'] = calendario_df['Data'].dt.dayofyear
    
    # Adicionar coluna com o mês e ano
    calendario_df['Mes_Ano'] = calendario_df['Data'].dt.to_period('M')
    
    return calendario_df

# Exemplo: criar um calendário para janeiro de 2023
ano = 2018
mes = 1
calendario_dim = calendario(ano, mes)

# Exibir o DataFrame
print(calendario_dim)
```

#### Transformando tabelas fatos e dimensão 
- Aqui eu me preocupei em criar tabelas dimensões sem duplicações e com o id identificador sendo os dados de estado correspondente as duas tabelas e a de cidade apenas a tabela de covid.
- OBS: Se novos regristros de estados ou cidades aparecerem em futuras etls o id deles podem mudar. 

```python
### Dimensão cidade tabela de covid

df_city_dim = df[(df['place_type'] == 'city')][['city','state']].drop_duplicates(subset=['city','state'])

### Criando ID_cidade

df_city_dim['id_city'] = range(1, len(df_city_dim) + 1)

### Dimensão estado da tabela covid
df_state_dim_cov = df[['state']].drop_duplicates(subset=['state'])

### Dimensão estado da tabela multas
df_state_dim_mul = df_multas[['uf']].drop_duplicates(subset=['uf']).rename(columns={'uf': 'state'})

### Juntando ambas as tabelas pela string
df_state_dim = pd.concat([df_state_dim_cov, df_state_dim_mul], ignore_index=True)

# Criando id estado 
df_state_dim['id_state'] = range(1, len(df_state_dim) + 1)

print(df_state_dim)
```

```python
##### Tabelas dimensão da tabela de multas

df_infracao = df_multas[['escopo_autuacao']].drop_duplicates(subset=['escopo_autuacao'])

##### Criando id
df_infracao['id_infracao'] = range(1, len(df_infracao) + 1)

##### Removendo descrição
df_multas = pd.merge(df_multas, df_infracao, how='left' , on= 'escopo_autuacao').drop(['escopo_autuacao'], axis=1) 
```

```python
df_multas
```

```python
### Fato covid removendo as decrições de cidade e estado e inserindo o id.
 
df = pd.merge(df, df_city_dim[['city','id_city']], how='left' , on= 'city').drop(['city'], axis=1)

df = pd.merge(df, df_state_dim[['state','id_state']], how='left' , on= 'state').drop(['state'], axis=1) 


#### Renomeando uf

df_multas.rename(columns={'uf': 'state'}, inplace=True)

### Subistituindo a descrição pelo id
df_multas = pd.merge(df_multas, df_state_dim[['state','id_state']], how='left' , on= 'state').drop(['state'], axis=1) 

### Criando campo de data
df_multas['data'] = df_multas['num_meses'].astype(str) + '/' + '1'  + '/' + df_multas['ano'].astype(str)

### drop de num meses 

df_multas = df_multas.drop(['num_meses'], axis=1)
```
